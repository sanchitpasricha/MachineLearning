## What is Gradient Descent

**Gradient Descent is an optimization algorithm for finding a local minimum of a differentiable function. Gradient descent in machine learning is simply used to find the values of a function's parameters (coefficients) that minimize a cost function as far as possible.**

```python
import numpy as np
```

```python
import numpy as np
import math

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 2000
    n = len(x)
    learning_rate = 0.01111
    tolerance = 1e-9  # Adjust this tolerance as needed
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost_function = (1/n) * sum([val**2 for val in (y - y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        expected_m = 2.0064773310697563
        expected_b = 2.9766147795921603
        expected_cost = 0.00010029166672711225
        expected_iteration = 1215
        
        is_m_close = math.isclose(m_curr, expected_m, rel_tol=tolerance)
        is_b_close = math.isclose(b_curr, expected_b, rel_tol=tolerance)
        is_cost_close = math.isclose(cost_function, expected_cost, rel_tol=tolerance)
        is_iteration_close = (i == expected_iteration)
        
        print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost_function, i))
        
        if is_m_close and is_b_close and is_cost_close and is_iteration_close:
            break
        

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
